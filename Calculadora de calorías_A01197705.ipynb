A continuación para programar la calculadora realizamos una recoleción de los datos de la manera que tomaremos en cuenta todos los contenidos nutricionales que implica el cálculo de calorías:

```python
name = input('Ingresa el nombre del alimento para calcular calorías: ' )
carb = float(input('Ingresa la cantidad de carbohidratos en gramos: '))   #Utilizamos funciones de input para recolectar los datos en cuestión 
lipidos = float(input('Ingresa la cantidad de lipidos en gramos:'))       #Utilizamos datos como float en caso de que la cantidad de gramos contengan decimales y así -
proteinas = float(input('Ingresa la cantidad de proteinas en gramos:'))   # - sean lo más precisos Posibles
```

Una vez recolectados los datos elaboramos una ecuación, basándonos en los coeficientes de regresión lineal que habiamos elaborado previamente en la actividad 3, que en mi caso particular fueron los siguientes: 


*   Carbohidratos:  3.7300749930919
*   Lípidos: 9.2706241330977
*   Proteinas: 3.97707523516352

Por lo que primero se definirán los valores de los coeficientes y posteriormente se desarrollara una función para poder ejecutar dicha función de la siguiente manera:






```python
#Definimos los valores de los coeficientes
CoefCarb = float(3.78085732963605)
CoefLipidos = float(9.10914604487066)
CoefProteinas = float(4.13994792482292)

#Realizamos función donde realizamos ecuación con los datos recolectados y los coeficientes para obtener las calorias

def calorias():
  getCalorias = ((carb*CoefCarb)+(lipidos*CoefLipidos)+(proteinas*CoefProteinas))
  print('Las calorías proporcionadas por una porción de ' + str(name) + ' son: ' )
  print(getCalorias)

calorias()
```
