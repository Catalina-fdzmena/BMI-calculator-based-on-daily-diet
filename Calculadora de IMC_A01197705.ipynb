Antes de realizar el cálculo es necesario el desarrollo de una lista para almacenar los datos necesarios para así obtener nuestro BMI/IMC, en dónde de hecho colocaré los datos propios.

```python
datos_persona = ['Katy','Fernandez',19,1.58,52] 
# Los datos de dicha lista corresponderán a: nombre, apellido, edad, estatura, peso
```

Hacemos la asignación múltiple de variables de la lista para usarlas de manera más sencilla en el código, en lugar usar índices en la lista:

```python
name, lastname, age, height, weight = datos_persona 
```

Una vez conociendo las variables definidas, se procede a hacer la fórmula para obtener el IMC 

```python
imc = weight / height ** 2   #estatura se coloca a la segunda potencia
```

Una vez realizando la prueba se realiza una impresión de los datos para así poder mostrar la información del programa

```python
print("BMI calculator")

print("¡Hi there,",name,lastname,"!")

print("Your weight registered is: ",weight,"kg")

print("Your height registered is: ",height,"m")

print("Acroding to this data collected, your BMI is the following  ",imc,"which indicates that: ")
```

Posterior a eso para reflejar qué tan saludable se encuetra tu IMC, procedemos a realizar una serie de condicionales para ver que va representar el valor del condicional indicado como tal

```python

if imc <= 18.4: # si el BMI es menor a 18.4
   print("Under weight")

elif imc >= 18.5 and imc <= 24.9: # sino, si el BMI es mayor a 18.5 y menor 24.9 
   print("Normal weight")

elif imc >= 25 and imc <= 29.9: # sino, si el IMC es mayor a 25 y menor a 29.9
   print("Overweight")

else: # sino cae en ninguno de los otros rangos entonces se asigna como obesidad
   print("Obesity")
```
